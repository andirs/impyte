## Sensoric Data

Larger file needs to be downloaded.

```python
import os
import urllib.request
```

```python
def load_sensoric_data():
    filename = "pml-training.csv"
    if not os.path.exists("../data/{}".format(filename)):
        print("Downloading file.")
        url = "https://www.dropbox.com/s/ukbdf12t0s29327/pml-training.csv?dl=1" 
        u = urllib.request.urlopen(url)
        data = u.read()
        u.close()
        with open("../data/{}".format(filename), "wb") as f :
            f.write(data)
    return pd.read_csv("../data/{}".format(filename), low_memory=False)
```

```python
sensoric_data = load_sensoric_data()
```

```python
reload(impyte)
imp = impyte.Impyter(sensoric_data)
```

```python
summa = imp.get_summary()
summa
```
